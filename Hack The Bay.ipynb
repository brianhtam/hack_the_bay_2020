## Do some Data Cleaning on selected_df

```python
# Strip white space from Parameter, ParameterName_CBP and ParameterName_CMC
selected_df['ParameterName_CBP'] = selected_df['ParameterName_CBP'].str.strip()
```

```python
selected_df['ParameterName_CMC'] = selected_df['ParameterName_CMC'].str.strip()
```

```python
selected_df['ParameterName_CMC'].unique()
```

```python
selected_df.query('ParameterName_CMC == "Chlorophyll a;b;c"  or ParameterName_CMC == "Chlorophyll a"')['ParameterName_CMC'].value_counts()
```

```python
# Standardize Parameter Values
# 'Dissolved oxygen % Sat' and 'Dissolved Oxygen % Sat' = 'Dissolved Oxygen % Sat'
# 'Air temperature' and 'Air Temperature' = 'Air Temperature'
# 'Water Temperature' and 'Water temperature' = 'Water Temperature'
# 'Chlorophyll a;b;c' and 'Chlorophyll a' = 'Chlorophyll A'
# Bacteria [E.Coli] = 'E.coli'

# Errorneous values
# Nitrite-nitrate = 'Nitrate-nitrogen'
```

```python
# Replace 'Nitrate-nitrate' with 'Nitrate-nitrogen'
selected_df['ParameterName_CMC'] = selected_df['ParameterName_CMC'].replace({
    'Nitrite-nitrate':'Nitrate-nitrogen',
    'Air temperature': 'Air Temperature',
    'Dissolved oxygen % Sat': 'Dissolved Oxygen % Sat',
    'Water temperature': 'Water Temperature',
    'Chlorophyll a;b;c': 'Chlorophyll A',
    'Chlorophyll a': 'Chlorophyll A'
    })
```

```python
selected_df['ParameterName_CMC'].unique()
```
